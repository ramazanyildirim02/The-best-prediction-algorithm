```python
# Gerekli kütüphaneleri içeri aktarıyoruz.
import numpy as np 
import pandas as pd
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
dataFrame = pd.read_csv("maaslar_yeni.csv")
dataFrame
```

## 1. UnvanSeviyesi, Kıdem, Puan değişkenlerini kullanarak maas tahmin etme

```python
X = dataFrame.iloc[:,2:5].values
Y = dataFrame.iloc[:,5:].values
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)

model=sm.OLS(lin_reg.predict(X),X).fit()
print(model.summary())
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
print(x_poly)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)

model2=sm.OLS(lin_reg2.predict(x_poly),X).fit()
print(model2.summary())
```

```python
#Verilerin olceklenmesi(scaling)
from sklearn.preprocessing import StandardScaler 

sca1er=StandardScaler()
x_olcekli = sca1er.fit_transform(X)
y_olcekli = sca1er.fit_transform(Y)

#Support Vector Regression
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
svr_reg.fit(x_olcekli,y_olcekli)

model3=sm.OLS(svr_reg.predict(x_olcekli),x_olcekli).fit()
print(model3.summary())
```

```python
#Decision Tree Regresyon
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=0)
tree_reg.fit(X,Y)

model4=sm.OLS(tree_reg.predict(X),X).fit()
print(model4.summary())
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators = 10,random_state=0)
rf_reg.fit(X,Y.ravel())

print('Random Forest OLS')
model5=sm.OLS(rf_reg.predict(X),X).fit()
print(model5.summary())
```

```python
r2_df1 = pd.DataFrame(data = [r2_score(Y, lin_reg.predict(X)),r2_score(Y, lin_reg2.predict(x_poly)),r2_score(y_olcekli, svr_reg.predict(x_olcekli)),
                             r2_score(Y, tree_reg.predict(X)),r2_score(Y, rf_reg.predict(X))],index = ["Linear","Polynomial","SVR","Decision Tree","Random Forest"],columns=["1.model R2"])
```

   p-value sonuçlarından da anlaşılabileceği üzere modeli sadece UnvanSeviyesi değişkeni kullanarak yapmak daha doğru olur.

## 2. UnvanSeviyesi değişkenini kullanarak maas tahmin etme


```python
X = dataFrame.iloc[:,2].values.reshape(-1,1) # Bir önceki model aksine sadece tek bir değişkeni alıyorum.
Y = dataFrame.iloc[:,5:].values.reshape(-1,1)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)

model=sm.OLS(lin_reg.predict(X),X).fit()
print(model.summary())
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(X)
print(x_poly)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)

model2=sm.OLS(lin_reg2.predict(x_poly),X).fit()
print(model2.summary())
```

```python
#Verilerin olceklenmesi(scaling)
from sklearn.preprocessing import StandardScaler 

sca1er=StandardScaler()
x_olcekli = sca1er.fit_transform(X)
y_olcekli = sca1er.fit_transform(Y)

#Support Vector Regression
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
svr_reg.fit(x_olcekli,y_olcekli)

model3=sm.OLS(svr_reg.predict(x_olcekli),x_olcekli).fit()
print(model3.summary())
```

```python
#Decision Tree Regresyon
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=0)
tree_reg.fit(X,Y)

model4=sm.OLS(tree_reg.predict(X),X).fit()
print(model4.summary())
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators = 10,random_state=0)
rf_reg.fit(X,Y.ravel())

print('Random Forest OLS')
model5=sm.OLS(rf_reg.predict(X),X).fit()
print(model5.summary())
```

```python
r2_df2 = pd.DataFrame(data = [r2_score(Y, lin_reg.predict(X)),r2_score(Y, lin_reg2.predict(x_poly)),r2_score(y_olcekli, svr_reg.predict(x_olcekli)),
                             r2_score(Y, tree_reg.predict(X)),r2_score(Y, rf_reg.predict(X))],index = ["Linear","Polynomial","SVR","Decision Tree","Random Forest"],columns=["2.model R2"])
```

```python
compare_df = pd.concat([r2_df1,r2_df2],axis=1)
compare_df
```

OLS sonuçlarındaki R2 değerleri ve r2_score() metodu kullanılarak çıkan R2 değerlerinin farklı olmasının sebebi R2'nin ikisinde farklı şekilde hesaplanmasıdır. Ama bu bizim yorum yapmamızı engellemez.

Bu veriler için tek değişkenli modelin daha doğru bir model olduğunu gördük. Fakat uzun uğraşlar yerine Korelasyon kullanarak bunu daha kısa bir şekilde görebiliriz.
